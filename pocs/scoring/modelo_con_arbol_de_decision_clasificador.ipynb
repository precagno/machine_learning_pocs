```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

model_data = pd.read_csv('dataset/train_clean.csv')
```

```python
model_data_clean = model_data[pd.notnull(model_data['Num_of_Delayed_Payment'])]
X = model_data_clean.drop(columns=['Credit_Score','ID','Month','Name','Payment_Behaviour'])
y = model_data_clean['Credit_Score']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model
```

```python
predictions = model.predict(X_test)

predictions
```

```python
score = accuracy_score(y_test, predictions)
score
```
